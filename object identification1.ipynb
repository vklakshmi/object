```python
df = pd.DataFrame(columns=['pos'])
```

```python
from nltk import pos_tag_sents,word_tokenize
```

```python
df['pos'] = pos_tag_sents(data['Final Narrative'].apply(word_tokenize))
```

```python
df['pos']
```

```python
grammar_text1= ("""
                NP:{<DT>?<JJ>*<NN>}
                P:{<IN>}
                V:{<V.*>}
                PP:{<P> <NP>}
                VP:{<V><NP|PP>*}
                """)
```

```python
from nltk import RegexpParser
```

```python
from nltk.tree import Tree
```

```python
chunker=RegexpParser(grammar_text1)
```

```python
for i in range(0,100):
    var=df['pos'].iloc[[i]]
    chunked=[]
    for s in var:
        chunked.append(chunker.parse(s))
    chunked_1 = ''.join(map(str,chunked))
    t=Tree.fromstring(chunked_1)
    for j in t.subtrees(filter = lambda x: x.label() =='VP'):
        for attributes in j.leaves():
            if attributes.endswith('/NN'):
                print attributes            
```

```python
noun=[]
for i in range(0,4933):
    var=df['pos'].iloc[[i]]
    chunked=[]
    for s in var:
        chunked.append(chunker.parse(s))
    chunked_1 = ''.join(map(str,chunked))
    t=Tree.fromstring(chunked_1)
    for j in t.subtrees(filter = lambda x: x.label() =='VP'):
        for attributes in j.leaves():
            if attributes.endswith('/NN'):
                noun.append(attributes) 
```

```python
from collections import defaultdict
```

```python
word_freq = defaultdict(int)
```

```python
for text in noun:
    for word in text.split():
        word_freq[word] += 1
```

```python
Sorted_Data=pd.DataFrame.from_dict(word_freq, orient='index') \
.sort_values(0, ascending=False) \
.rename(columns={0: 'abs_freq'})
```

```python
Sorted_Data
```

```python
Sorted_Data.drop(Sorted_Data[Sorted_Data.abs_freq > 760].index, inplace=True)
```
